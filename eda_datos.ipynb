```python
for col in df.columns:
    # Create barplot directly from the DataFrame
    sns.countplot(data=df, x=col, palette='Blues')

    # Add title
    plt.title(f'{col} Counts')

    # Show the plot
    plt.show()
```

```python
plt.figure(figsize = (10, 30))
for i, col in enumerate(df.select_dtypes(exclude=['object']).columns, 1):
    plt.subplot(9, 2, i)
    sns.boxplot(data=df, x=col, y="sex")
    plt.title(f"Boxplot of {col} vs sex")
    plt.tight_layout()
    plt.plot()
```

```python
plt.figure(figsize = (10, 30))

for i, col in enumerate(df.select_dtypes(exclude=['object']).columns, 1):
    if df[col].dtype == object: # se ignoran las variables no numericas
        continue
    plt.subplot(9, 2, i)
    skewness = df[col].skew()
    sns.distplot(df[col], label = "Skew = %.3f" %(skewness))
    plt.title(f"Skewness of {col} Data")
    plt.tight_layout()
    plt.legend(loc = "best")
    plt.plot()
```
